# Outline

Some fractals are easy to draw. For the <a href="http://en.wikipedia.org/wiki/Mandelbrot_set">Mandelbrot Set</a>, there is a condition which can be evaluated (or at least approximated with sufficient accuracy) at each point (i.e. pixel) which determines whether it should be coloured or not. So we can just loop through the pixels once and test each one.

Other times, there is no such simple rule. We just have to keep drawing until the elements become small enough not to have any visual impact. This happens when they're a bit smaller than a pixel. That's what we'll be doing here. For the Apollonian Gasket and Sierpinski triangle, there is only one type of element to draw - circles or triangles - and there is just one rule by which they're drawn. Again, no such luck here. We have both triangles and circles. What's more, there are three subtely different types of circle:

<ol>
 <li>The first one, which sits snugly inside the triangle.</li>
 <li> Ones which sit between two of the straight lines of the triangle and an existing circle.</li>
 <li> Ones which sit between one of the straight lines of the triangle and two existing circles.</li>
 <li> Ones which sit between three existing circles.</li>
</ol> 

Each of these will need to be handle slightly differently.

# Basic Elements

Our fractal will be made of two sorts of element, circles and line segments. We don't need full blown classes for these, but some sort of data structure would be good. Named tuples will do nicely. Circles are defined by a centre and a radius. Lines are defined by two end points.

```python
from collections import namedtuple

Circle = namedtuple('Circle', ['cen', 'rad'])
Line   = namedtuple('Line', ['end1', 'end2'])
```

While we're here, lets make some functions for drawing them.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def draw_line(ax, line, colour='b', linewidth=2):
    """
    Draw a line object on an axis
    """
    ax.plot([line.end1[0],line.end2[0]],[line.end1[1],line.end2[1]],color=colour,linewidth=linewidth)
    
    return


def draw_circle(ax, circle, colour='b', linewidth=2):
    """
    Draw a circle object on an axis
    """
    circle_artist = plt.Circle(circle.cen,circle.rad,color=colour,fill=False,linewidth=linewidth)
    ax.add_artist(circle_artist)
    
    return
```

# Soddy Circles

## Type 4 Circles

We'll start with the fourth type of circle. Suppose we have three mutually tangent circles, we want to find one which sits in the middle, mutually tangent to all three. For example:

```python
import numpy as np

# Make three mutually tangent circles
circle1 = Circle((-1,0),1)
circle2 = Circle((1,0),1)
circle3 = Circle((0,np.sqrt(3)),1)

# Take a stab at where the inner circle should go
circle4 = Circle((0,0.65),0.15)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_circle(ax, circle1)
draw_circle(ax, circle2)
draw_circle(ax, circle3)
draw_circle(ax, circle4, colour='r')

plt.show()
```

Missed! We need a more principled way to work out where the inner circle should go. It turns out that there are some extraorinarily neat formulas for calculating these. The radiuses of the four circles are linked by <a href="http://en.wikipedia.org/wiki/Descartes%27_theorem">Descartes's Circle Theorem</a>. This theorem has been discovered and rediscovered multiple times, but extra credit is due to chap called Frederick Soddy, because he published the result in <em>Nature</em> in the form of a poem, called The Kiss Precise. Here's the first two verses.

For pairs of lips to kiss maybe<br/>
Involves no trigonometry.<br/>
This not so when four circles kiss<br/>
Each one the other three.<br/>
To bring this off the four must be<br/>
As three in one or one in three.<br/>
If one in three, beyond a doubt<br/>
Each gets three kisses from without.<br/>
If three in one, then is that one<br/>
Thrice kissed internally.<br/>

Four circles to the kissing come.<br/>
The smaller are the benter.<br/>
The bend is just the inverse of<br/>
The distance form the center.<br/>
Though their intrigue left Euclid dumb<br/>
There's now no need for rule of thumb.<br/>
Since zero bend's a dead straight line<br/>
And concave bends have minus sign,<br/>
The sum of the squares of all four bends<br/>
Is half the square of their sum.<br/>

<em>Frederick Soddy, in Nature, June 20, 1936, borrowed from <a href="http://www.pballew.net/soddy.html">here</a></em>

To summarise, if our three outer circles have radiuses $r_i$, $i = 1,2,3$, then we can define their <em>bends</em> as $b_i=\frac{1}{r_i}$, $i = 1,2,3$. Let the Soddy circle have radius and bend $r_s$ and $b_s$, then,

\begin{equation}
(b_1 + b_2 + b_3 + b_s)^2 = 2 (b_1^2 + b_2^2 + b_3^2 + b_s^2)     .
\end{equation}

This has two solutions,

\begin{equation}
b_s = b_1 + b_2 + b_3 \pm 2 \sqrt{ b_1 b_2 + b_2 b_3 + b_3 b_1 }     .
\end{equation}

As the poem mentions, there are two Soddy circles, the little one that we want, and a big one which surrounds the three parent circles. The little one is the positive solution. Considering how elegant and concise this relationship is, the proof is surprisingly tedious. However there's nothing too conceptually difficult about it. It's in <a href="http://www.jstor.org/stable/2315097">this paper</a> (for which you'll need a JSTOR subscription), and there's also a nice version <a href="http://euler.genepeer.com/from-herons-formula-to-descartes-circle-theorem/">here</a> (which is free).

That takes cares of the radius. What about the centre? It turns out that this is given by <em>the same formula</em>. Suppose that our three parent circles have centres at $(x_i,y_i)$, $i = 1,2,3$. If we write this as a complex number $c_i = x_i + j y_i$ (with $j = \sqrt{-1}$), then define $z_i = c_i b_i$, suddenly we find,

\begin{equation}
(z_1 + z_2 + z_3 + z_s)^2 = 2 (z_1^2 + z_2^2 + z_3^2 + z_2^2)     .
\end{equation}

That's really neat, right? It was published in <a href="http://arxiv.org/abs/math/0101066">this paper</a>, but there's a more comprehensible proof <a href="http://faculty.plattsburgh.edu/sam.northshield/DescCThmV4.pdf">here</a>.  I still haven't got my head round this. As before there are two solutions,

\begin{equation}
z_s = z_1 + z_2 + z_3 \pm 2 \sqrt{ z_1 z_2 + z_2 z_3 + z_3 z_1 }     .
\end{equation}

Unfortunately, this time the one we want is not always the positive one. Until I have time to read that paper in more detail, we'll just have to make do with seeing which of the two possible centres fits best. We can do this by calculating the difference between the centre positions and comparing with the sum of the radiuses. Pick whichever choice minimises this error.

We've now got everything we need to find the new circle. Here's the function we need.

```python
def circle_in_three_circles(circ1, circ2, circ3):
    """
    Find the small Soddy circle for three existing circles
    """
    
    # Bends
    b1 = 1./circ1.rad
    b2 = 1./circ2.rad
    b3 = 1./circ3.rad
    
    # Solve Descartes circle theorem
    bs = b1 + b2 + b3 + 2*np.sqrt( b1*b2 + b2*b3 + b3*b1 )
    radius = 1./bs
    
    # Centre-bend products
    z1 = b1 * complex(circ1.cen[0],circ1.cen[1])
    z2 = b2 * complex(circ2.cen[0],circ2.cen[1])
    z3 = b3 * complex(circ3.cen[0],circ3.cen[1])
    
    # Solve complex Descartes circle theorem - two solutions
    zspos = z1 + z2 + z3 + 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    zsneg = z1 + z2 + z3 - 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    
    # Possible centres
    centrepos = zspos/bs
    centreneg = zsneg/bs
    
    # See which one is closer to fitting 
    errpos = abs(np.linalg.norm(np.array(circ1.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ1.rad)) \
           + abs(np.linalg.norm(np.array(circ2.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ2.rad)) \
           + abs(np.linalg.norm(np.array(circ3.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ3.rad))
    errneg = abs(np.linalg.norm(np.array(circ1.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ1.rad)) \
           + abs(np.linalg.norm(np.array(circ2.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ2.rad)) \
           + abs(np.linalg.norm(np.array(circ3.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ3.rad))
    
    # Choose one
    if errneg > errpos:
        centre = centrepos
    else:
        centre = centreneg
    
    # Make a circle object
    soddy_circle = Circle((centre.real,centre.imag),radius)
    
    return soddy_circle 
```

Let's try that out.

```python
# Make three mutually tangent circles
circle1 = Circle((-1,0),1)
circle2 = Circle((1,0),1)
circle3 = Circle((0,np.sqrt(3)),1)

# Soddy circle
circle4 = circle_in_three_circles(circle1,circle2,circle3)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_circle(ax, circle1)
draw_circle(ax, circle2)
draw_circle(ax, circle3)
draw_circle(ax, circle4, colour='r')

plt.show()
```

Bingo. That looks good.

## Type 3 Circles

Now to an edge case, when instead of one of the circles we have an edge. Like this:

```python
# Make some circles...
circle1 = Circle((-1,0),1)
circle2 = Circle((1,0),1)

# ... and a line
line = Line((-2,1),(2,1))

# Take a guess at where the inner circle should go
circle3 = Circle((0,0.75),0.2)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_circle(ax, circle1)
draw_circle(ax, circle2)
draw_line(ax, line)
draw_circle(ax, circle3, colour='r')

plt.show()
```

Missed again. We don't need much extra here. That straight line is just like a circle with an infinitely large radius, and thus the corresponding bend is $0$. Plug that into Descartes's Circle Theorem and we've got the bend.

\begin{equation}
b_s = b_1 + b_2 + 2 \sqrt{ b_1 b_2 }     .
\end{equation}

The centre is a little harder, and yet really neat. The centre of the circle is somewhere off in the direction perpendicular to the straight line. Let's draw a unit vector in this direction and call it $u$, like this:

```python
ax.arrow(-1, 1.5, 0, 1, head_width=0.1, head_length=0.1, fc='g', ec='g')
ax.text(-0.9, 2, '$u$', size=20)
fig
```

Now, suppose the radius of that really big circle is $R$, then the centre is at $a + jb + R(u_x + j u_y)$, where $a+jb$ is the point where the really big circle and the Soddy circle touch. The neat thing is that it doen't matter where this point is. The centre-bend product for the really big circle is,

\begin{equation}
z_{\text{big}} = \lim_{R\to\infty} \frac{a + jb + R(u_x + j u_y)}{R} = u_x + j u_y      .
\end{equation}

We can plug that into the complex from of Descartes's circle theorem,

\begin{equation}
z_s = z_1 + z_2 + z_{\text{big}} \pm 2 \sqrt{ z_1 z_2 + z_2 z_{\text{big}} + z_{\text{big}} z_1 }     .
\end{equation}

Let's code it up.




```python
def circle_in_two_circles_and_a_line(circ1, circ2, line):
    """
    Find the Soddy circle for two existing circles and a tangent line
    """
    
    # Bends
    b1 = 1./circ1.rad
    b2 = 1./circ2.rad
    
    # Solve Descartes circle theorem
    bs = b1 + b2 + 2*np.sqrt( b1*b2 )
    radius = 1./bs
    
    # Centre-bend products
    z1 = b1 * complex(circ1.cen[0],circ1.cen[1])
    z2 = b2 * complex(circ2.cen[0],circ2.cen[1])
    
    # Limiting centre-bend
    vec = np.array(line.end2)-np.array(line.end1)
    unit_vec = vec/np.linalg.norm(vec)
    z3 = complex(-unit_vec[1],unit_vec[0])
    
    # Solve complex Descartes circle theorem - two solutions
    zspos = z1 + z2 + z3 + 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    zsneg = z1 + z2 + z3 - 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    
    # Possible centres
    centrepos = zspos/bs
    centreneg = zsneg/bs
    
    # See which one is closer to fitting 
    errpos = abs(np.linalg.norm(np.array(circ1.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ1.rad)) \
           + abs(np.linalg.norm(np.array(circ2.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ2.rad))
    errneg = abs(np.linalg.norm(np.array(circ1.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ1.rad)) \
           + abs(np.linalg.norm(np.array(circ2.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ2.rad))
    
    # Choose one
    if errneg > errpos:
        centre = centrepos
    else:
        centre = centreneg
    
    # Make a circle object
    soddy_circle = Circle((centre.real,centre.imag),radius)
    
    return soddy_circle 
```

```python
# Make some circles...
circle1 = Circle((-1,0),1)
circle2 = Circle((1,0),1)

# ... and a line
line = Line((-2,1),(2,1))

# Find the Soddy circle
circle3 = circle_in_two_circles_and_a_line(circle1, circle2, line)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_circle(ax, circle1)
draw_circle(ax, circle2)
draw_line(ax, line)
draw_circle(ax, circle3, colour='r')

plt.show()
```

There's something we need to beware of here, the sign of the unit vector. If we switch the two end points of the line segment, then this vector will point the other way, and the Soddy circle will be in the wrong place. To make things work out, we need to make sure that triangles all go the same way round. Here we're using the convention that they go clockwise. We'll need to enforce this condition later when we get onto defining triangles.

# Incircles

## Type 1 Circles

The first circle we draw is unique. It sits inside the original triangle, with the three sides all tangent to it. This is called the incircle of the triangle. 

```python
# Make three lines which comprise a triangle
line1 = Line((-1,-1),(0.5,2))
line2 = Line((0.5,2),(1,-1))
line3 = Line((1,-1),(-1,-1))

# Take a guess at where the incircle should go
incircle = Circle((0.3,0),0.5)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_line(ax, line1)
draw_line(ax, line2)
draw_line(ax, line3)
draw_circle(ax, incircle, colour='r')

plt.show()
```

Curses, way off again. <a href="http://en.wikipedia.org/wiki/Incircle_and_excircles_of_a_triangle">Wikipedia</a> lists all sorts of useful properties of incircles including the ones we need, its centre and radius. Specifically, if the three corners have coordinates $(x_a,y_a)$, $(x_b,y_b)$, $(x_c,y_c)$, and the opposite sides have lengths $l_a$, $l_b$, $l_c$, then the radius of the incircle is,

\begin{equation}
r_I = \sqrt{ \frac{(s-l_a)(s-l_b)(s-l_c)}{s} }     ,
\end{equation}

where $s$ is the <em>semiperimeter</em>, $s = \frac{1}{2}(l_a + l_b + l_c)$, the coordinates of the centre of the incircle (imaginatively named the <em>incentre</em>) are,

\begin{array}{c}
 x_I = \frac{l_a x_a + l_b x_b + l_c x_c}{2s} \\
 y_I = \frac{l_a y_a + l_b y_b + l_c y_c}{2s}      .
\end{array}

Let's try that out.

```python
def circle_in_three_lines(line1, line2, line3):
    """
    Find the incircle of a triangle specified by three lines
    """
    
    # End point coordinates
    xa = line1.end1[0]
    ya = line1.end1[1]
    xb = line2.end1[0]
    yb = line2.end1[1]
    xc = line3.end1[0]
    yc = line3.end1[1]
    
    # Lengths
    la = np.linalg.norm(np.array([xc-xb,yc-yb]))
    lb = np.linalg.norm(np.array([xa-xc,ya-yc]))
    lc = np.linalg.norm(np.array([xb-xa,yb-ya]))
    
    # Semiperimeter
    s = 0.5*(la+lb+lc)
    
    # Incircle radius
    ri = np.sqrt((s-la)*(s-lb)*(s-lc)/s)
    
    # Incentre
    xi = (la*xa + lb*xb + lc*xc)/(2*s)
    yi = (la*ya + lb*yb + lc*yc)/(2*s)
    
    # Make incircle
    incircle = Circle((xi,yi),ri)
    
    return incircle
    
```

```python
# Make three lines which comprise a triangle
line1 = Line((-1,-1),(0.5,2))
line2 = Line((0.5,2),(1,-1))
line3 = Line((1,-1),(-1,-1))

# Find the incircle
incircle = circle_in_three_lines(line1,line2,line3)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_line(ax, line1)
draw_line(ax, line2)
draw_line(ax, line3)
draw_circle(ax, incircle, colour='r')

plt.show()
```

Lovely.

## Type 2 Circles

The last sort of circle that we'll come come across is those which fill in the corners, fitting in between two converging lines of the primary triangle and one existing circle. Like this:

```python
# Make two lines tangent to one circle
line1 = Line((-2*np.sqrt(3)/2,-1),(0,2))
line2 = Line((0,2),(2*np.sqrt(3)/2,-1))
circle = Circle((0,0),1)

# Take a guess at where the incircle should go
incircle = Circle((0,1.35),0.3)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_line(ax, line1)
draw_line(ax, line2)
draw_circle(ax, circle)
draw_circle(ax, incircle, colour='r')

plt.show()
```

Guessing is not the way. What we're trying to draw here is really just another incircle. Let's add a helpful line which is tangent to our circle and perpendicular to the direction from the centre to the corner made by the two lines.

```python
fig.gca().plot([-2,2],[1,1],'g',linewidth=2)
fig
```

We need to draw the incircle of the isosceles triangle formed by the green line and the two principal lines of the triangle. So we can just reuse our existing code for that. The only additional work we need to do is find the corners of the triangle, where the green line intersects with the blue ones. This requires some fiddly geometry. Peer at this construction for a bit.

```python
# Set up
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

# Draw the basic set up
line1 = Line((-2*np.sqrt(3)/2,-1),(0,2))
line2 = Line((0,2),(2*np.sqrt(3)/2,-1))
circle = Circle((0,0),1)
draw_line(ax, line1)
draw_line(ax, line2)
draw_circle(ax, circle)

# Add construction lines
helper_line = Line((-2,1),(2,1))
construction_line1 = Line((0,0),(0,2))
construction_line2 = Line((0,0),(np.sqrt(3)/2,0.5))
draw_line(ax, helper_line, colour='g')
draw_line(ax, construction_line1, colour='m')
draw_line(ax, construction_line2, colour='m')

# Add labels
ax.text(-0.1, 2.1, '$A$', size=20)
ax.text(0.9, 0.5, '$B$', size=20)
ax.text(-0.3, -0.2, '$C$', size=20)
ax.text(0.6, 1.1, '$D$', size=20)
ax.text(-0.3, 1.1, '$E$', size=20)
ax.text(-0.2, 0.4, '$r$', size=20)
ax.text(0.4, 0, '$r$', size=20)
ax.text(0.02, 1.35, '$x$', size=20)
ax.text(0.2, 1.1, '$y$', size=20)

# Add the unit vectors
ax.arrow(-1, 1.5, 0, 1, head_width=0.1, head_length=0.1, fc='k', ec='k')
ax.text(-0.9, 2, '$u$', size=20)
ax.arrow(-1, 1.5, -1, 0, head_width=0.1, head_length=0.1, fc='k', ec='k')
ax.text(-1.6, 1.6, '$v$', size=20)

plt.show()
```

We'll write position vectors the points as $p_A$, $p_B$, etc. The things we know to begin with are $p_A$, $p_C$ and $r$.

First we find $u$, a unit vector in the direction from $C$ towards $A$. That's going to be,

\begin{equation}
u = \frac{ p_A-p_C }{ \left|p_A-p_C\right| }      .
\end{equation}

At the same time the length of the vector gives us,

\begin{equation}
x = \left|p_A-p_C\right| - r      .
\end{equation}

If we rotate $u$ by $90^{\circ}$ then we get another unit vector which points along the green line. We'll need that later.

\begin{equation}
v = \begin{bmatrix}0 & -1 \\ 1 & 0\end{bmatrix} u
\end{equation}

The position of point $E$ is now just $p_E = p_C + r u$.

We can find the length $y$ using similar triangles. $ABC$ is similar to $AED$, which means, along with a little Pythagoras,

\begin{eqnarray}
\frac{y}{x} &= \frac{r}{\sqrt{(r+x)^2-r^2}} \\
\Rightarrow y &= \frac{r}{\sqrt{1+2\frac{r}{x}}}     .
\end{eqnarray}

Finally, point D is at $p_D = p_E - y v$. The other end is easily obtained by symmetry. Coding time!

```python
def circle_in_two_lines_and_a_circle(line1, line2, circle):
    """
    Find the incircle between two converging lines and a tangent circle
    """
    
    # See if the lines need switching round
    if line2.end1 != line1.end2:
        tmpline = line2
        line2 = line1
        line1 = tmpline
    
    # Get the centre and radius of the circle and the vertex where the lines meet
    point = np.array(line1.end2)
    centre = np.array(circle.cen)
    radius = np.array(circle.rad)
    
    # Find a unit vector pointing from the centre to the vertex
    vector1 = point-centre
    vectorlength = np.linalg.norm(vector1)
    vector1 = vector1/vectorlength
    
    # Find the intersection point where the helper line is tangent to the circle
    intersect = centre + radius*vector1
    
    # Find the half-length of the helper line
    halflength = radius/np.sqrt( 1 + 2*radius/(vectorlength-radius) )
    
    # Rotate the unit vector so it points along the helper line
    vector2 = np.dot(np.array([[0,-1],[1,0]]),vector1)
    
    # Find the ends of the helper line
    end1 = tuple(intersect - halflength*vector2)
    end2 = tuple(intersect + halflength*vector2)
    helper_line = Line(end1,end2)
    
    # Make shortened copies of the other two lines
    shortline1 = Line(end2, line1.end2)
    shortline2 = Line(line2.end1, end1)
    
    # Call the existing incircle routine
    incircle = circle_in_three_lines(shortline1, shortline2, helper_line)
    
    return incircle
    
```

```python
# Make two lines tangent to one circle
line1 = Line((-2*np.sqrt(3)/2,-1),(0,2))
line2 = Line((0,2),(2*np.sqrt(3)/2,-1))
circle = Circle((0,0),1)

# Find the incircle
incircle = circle_in_two_lines_and_a_circle(line1, line2, circle)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_line(ax, line1)
draw_line(ax, line2)
draw_circle(ax, circle)
draw_circle(ax, incircle, colour='r')

plt.show()
```

That's it. No more geometry. We have all the routines we need to add each different type of circle. Now it's recursion time.


# Building the Fractal

The input to our fractal circle routine will be a triangle. The first thing we'll need is function to parse this triangle and put it into the correct form. A sensible way to input a triangle is as three sets of 2D coordinates for the corners. No redundancy there. But our routines require it as a set of three lines. What's more, we need to make sure those lines are going clockwise, because of the rotations in the two-circles-and-a-line and two-lines-and-a-circle functions. A quick way to achieve this is to first guess, assigning the three sides in a random order, then evaluate the cross product of side 1 and side 2. The cross product is right-handed, so if the out-of-plane component is negative then we've got it correct. Here goes...

```python
class TriangleError(ValueError):
    """
    Invalid triangle specification
    """

def convert_triangle(triangle):
    """
    Convert a list of three 2D points representing a triangle into a set of sides.
    Raise an error if the triangle is not a triangle.
    """
    
    # Make sure it has exactly three corners
    if len(triangle) != 3:
        raise TriangleError("Triangle must have three corners")
    
    # Get corners
    a = triangle[0]
    b = triangle[1]
    c = triangle[2]
    
    # Make sure each corner has two coordinates
    if ((len(a)!=2) or (len(b)!=2) or (len(c)!=2)):
        raise TriangleError("Corners must have two coordinates")
        
    try:
        # Make sides from corners, taking care of the sign
        if ( (b[0]-a[0])*(c[1]-b[1]) - (b[1]-a[1])*(c[0]-b[0]) ) < 0:
            line1 = Line(a,b)
            line2 = Line(b,c)
            line3 = Line(c,a)
        else:
            line1 = Line(a,c)
            line2 = Line(c,b)
            line3 = Line(b,a)
            
    except TypeError:
        raise TriangleError("Corners are not valid numbers")
        

    sides = [line1,line2,line3]
    
    return sides
```

We're ready to start on the main fractal building routine. This will use two lists. One will be a list of all the things that comprise our fractal (three lines and a shed load of circles), each indexed by its position in the list. The second list is a queue of things to do. Each element of this list will be a tuple of indexes pointing to three things which could be parents of a new circle. I like the idea of drawing circles breadth-first, so we'll be popping items off the to-do list from the left. This is inefficient with a standard python list, so we'll need a deque instead.

Each time we pop an item off the queue, we just need to check how many of the parents are lines, then call the appropriate function to add a new circle. Each new circle spawns three new possibilities which are added to the to-do list, but only if the new circle is above some radius threshold. It's nice if we always keep the to-do triples in numerical order since it makes it easier to check how many are line.

That's it. Pretty simple. Here's the code.

```python
from collections import deque

def pack_circles_in_triangle(triangle, radius_limit):
    """
    Recursively pack circles into a triangle
    """
    
    # Convert triangle (3 points) into a set of side
    sides = convert_triangle(triangle)
    
    # Create a circle list
    circle_list = list(sides)
    counter = 2
    
    # Create a to-do stack
    todo_list = deque([(0,1,2)])
    
    # Loop
    it = 0
    while todo_list:
        it += 1
        
        # What's next
        parents = todo_list.popleft()
        
        # How many lines are present
        num_lines = 0
        for ii in range(3):
            if parents[ii] < 3:
                num_lines += 1
        
        # Switch on number of lines
        if num_lines == 0:
            
            # Three circles
            circle1 = circle_list[parents[0]]
            circle2 = circle_list[parents[1]]
            circle3 = circle_list[parents[2]]
            
            # Spawn a new circle
            new_circle = circle_in_three_circles(circle1, circle2, circle3)
            
        elif num_lines == 1:
            
            # Two circles, one line
            line1 = circle_list[parents[0]]
            circle1 = circle_list[parents[1]]
            circle2 = circle_list[parents[2]]
            
            # Spawn a new circle
            new_circle = circle_in_two_circles_and_a_line(circle1, circle2, line1)
            
        elif num_lines == 2:
            
            # Two lines, one circle
            line1 = circle_list[parents[0]]
            line2 = circle_list[parents[1]]
            circle1 = circle_list[parents[2]]
            
            # Spawn a new circle
            new_circle = circle_in_two_lines_and_a_circle(line1, line2, circle1)
            
        elif num_lines == 3:
            
            # Three line
            line1 = circle_list[parents[0]]
            line2 = circle_list[parents[1]]
            line3 = circle_list[parents[2]]
            
            # Spawn a new circle
            new_circle = circle_in_three_lines(line1, line2, line3)
            
        else:
            raise ValueError
        
        if new_circle.rad > radius_limit:
        
            # Add the new circle to the list
            circle_list.append(new_circle)
            counter += 1
            
            # Add a new parent combinations to the to-do stack
            todo_list.append((parents[0],parents[1],counter))
            todo_list.append((parents[0],parents[2],counter))
            todo_list.append((parents[1],parents[2],counter))
    
    
    # Remove the three lines from the list
    circle_list[:3] = []
    
    return sides,circle_list
    
```

Finally we want to draw the thing. We'll create a little wrapper which does this just by looping down the list.

```python
def draw_triangle_fractal(ax, triangle, radius_limit, linecolour='k', circlecolour='b', linewidth=2):
    """
    Draw a triangle and pack circles in it.
    """
    sides,circle_list = pack_circles_in_triangle(triangle, radius_limit)
    
    for ii in range(3):
        draw_line(ax, sides[ii], colour=linecolour, linewidth=linewidth)
    
    for circle in circle_list:
        draw_circle(ax, circle, colour=circlecolour, linewidth=linewidth)
    
    return
```

All done. Let's try it out.

# An Apollonian Addendum

Ok. Ok. I know. We've got everything we need to draw Apollonian Gaskets. Let's do that while we're here.

An Apollonian Gasket starts with three mutually tangent circles, one big one and two inside. Then we just draw soddy circles in all the gaps. So we want something like this:

```python
# Make three mutually tangent circles with which to start an Apollonian Gasket
circle1 = Circle((-1,0),1)
circle2 = Circle((1,0),1)
circle3 = Circle((0,0),2)

# Plot it all
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))

draw_circle(ax, circle1)
draw_circle(ax, circle2)
draw_circle(ax, circle3)

plt.show()
```

A minimum, non-redundant specification of the problem is the centres of the two smaller circles and the ratio of their radiuses. The first thing we need is a function to parse that and return the three circles.

```python
def convert_circles(centre1, centre2, radius_ratio):
    """
    Create thee circles with which to start the Apollonian Gasket from a minimum spec.
    """
    
    # Make sure the radiuses are both positive
    if (radius_ratio<0) or (radius_ratio>1):
        raise ValueError("Radius ratio must be between 0 and 1.")
        
    # Make sure each centre has two coordinates
    if ((len(centre1)!=2) or (len(centre2)!=2)):
        raise TriangleError("Corners must have two coordinates")
    
    # Make numpy vectors for both centres
    c1_vector = np.array(centre1)
    c2_vector = np.array(centre2)
    
    # Find the centre of the big circle
    c3_vector = c1_vector*radius_ratio + c2_vector*(1-radius_ratio)
    
    # Find the radius of the big circle
    radius3 = np.linalg.norm(c2_vector-c1_vector)
    
    # Find the radiuses of the little circles
    radius1 = radius3*radius_ratio
    radius2 = radius3*(1-radius_ratio)
    
    # Make circles
    circle1 = Circle(tuple(centre1),radius1)
    circle2 = Circle(tuple(centre2),radius2)
    circle3 = Circle(tuple(c3_vector),radius3)
    
    return [circle3,circle1,circle2]
```

Let's make sure this works.

Now when I said we had everything we needed for the Apolonian Gasket, that wasn't quite true. When one of our three circles is the big one, then we're not drawing a Soddy circle surround by three others. The Soddy circle in <em>inside</em> one of them. This means, as Soddy tells us in his poem, that the bend should be negative.

Furthermore, from the initial set up, we actually want <em>both</em> solutions to the complex theorem, since there will be two circles (of the same radius) to draw, one on each side. In this case, it's pot luck which centre our basic algorithm will return. We can just stick in an extra switch allowing us to ask for the other one.

We'll assume the all-encosing circle is always the first argument entered.

```python
def circle_in_three_circles_with_one_inside_out(circ1, circ2, circ3, other_one=False):
    """
    Find the two Soddy circles for three existing circles, where the first one encloses all the others.
    """
    
    # Bends
    b1 = -1./circ1.rad
    b2 = 1./circ2.rad
    b3 = 1./circ3.rad
    
    # Solve Descartes circle theorem
    bs = b1 + b2 + b3 + 2*np.sqrt( b1*b2 + b2*b3 + b3*b1 )
    radius = 1./bs
    
    # Centre-bend products
    z1 = b1 * complex(circ1.cen[0],circ1.cen[1])
    z2 = b2 * complex(circ2.cen[0],circ2.cen[1])
    z3 = b3 * complex(circ3.cen[0],circ3.cen[1])
    
    # Solve complex Descartes circle theorem - two solutions
    zspos = z1 + z2 + z3 + 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    zsneg = z1 + z2 + z3 - 2*np.sqrt( z1*z2 + z2*z3 + z3*z1 )
    
    # Possible centres
    centrepos = zspos/bs
    centreneg = zsneg/bs
    
    # See which one is closer to fitting 
    errpos = abs(np.linalg.norm(np.array(circ2.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ2.rad)) \
           + abs(np.linalg.norm(np.array(circ3.cen)-np.array([centrepos.real,centrepos.imag]))-(radius+circ3.rad))
    errneg = abs(np.linalg.norm(np.array(circ2.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ2.rad)) \
           + abs(np.linalg.norm(np.array(circ3.cen)-np.array([centreneg.real,centreneg.imag]))-(radius+circ3.rad))
    
    # Choose one
    if not other_one:
        if errneg > errpos:
            centre = centrepos
        else:
            centre = centreneg    
    else:
        if errneg > errpos:
            centre = centreneq
        else:
            centre = centrepos
    
    # Make a circle object
    soddy_circle = Circle((centre.real,centre.imag),radius)
    
    return soddy_circle 
```

Good. Time to copy and paste the main fractal loop function.

```python
def pack_circles_in_circle(centre1, centre2, radius_ratio, radius_limit):
    """
    Recursively pack circles into a circle
    """
    
    # Convert the input parameters into three circles
    initial_circles = convert_circles(centre1, centre2, radius_ratio)
    
    # Create a circle list
    circle_list = list(initial_circles)
    
    # Handle the first iteration specially
    new_circle1 = circle_in_three_circles_with_one_inside_out(circle_list[0], circle_list[1], circle_list[2])
    new_circle2 = circle_in_three_circles_with_one_inside_out(circle_list[0], circle_list[1], circle_list[2],
                                                              other_one=True)
    circle_list.extend([new_circle1,new_circle2])
    
    # Create a to-do stack
    todo_list = deque([(0,1,3),(0,2,3),(1,2,3),(0,1,4),(0,2,4),(1,2,4)])
    counter = 4
    
    # Loop
    it = 0
    while todo_list:
        it += 1
        
        # What's next
        parents = todo_list.popleft()
        
        # Three circles
        circle1 = circle_list[parents[0]]
        circle2 = circle_list[parents[1]]
        circle3 = circle_list[parents[2]]
        
        # Is the first circle the big one?
        if parents[0]==0:
            
            # Modified Soddy circles
            new_circle = circle_in_three_circles_with_one_inside_out(circle1, circle2, circle3)
            
        else:
            
            # Ordinary Soddy circle
            new_circle = circle_in_three_circles(circle1, circle2, circle3)
        
        if new_circle.rad > radius_limit:
        
            # Add the new circle to the list
            circle_list.append(new_circle)
            counter += 1
            
            # Add a new parent combinations to the to-do stack
            todo_list.append((parents[0],parents[1],counter))
            todo_list.append((parents[0],parents[2],counter))
            todo_list.append((parents[1],parents[2],counter))
    
    return circle_list
```

```python
def draw_circle_fractal(ax, centre1, centre2, radius_ratio, radius_limit,
                        bordercolour='k', circlecolour='b', linewidth=2):
    """
    Draw a circle and pack circles in it.
    """
    circle_list = pack_circles_in_circle(centre1, centre2, radius_ratio, radius_limit)
    
    draw_circle(ax, circle_list[0], colour=bordercolour, linewidth=linewidth)
    for circle in circle_list[1:]:
        draw_circle(ax, circle, colour=circlecolour, linewidth=linewidth)
    
    return
```
